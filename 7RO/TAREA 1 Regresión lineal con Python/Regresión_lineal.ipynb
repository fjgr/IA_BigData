```python
# Cálculo realizado con sklearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Datos obtenidos de la tabla
xi = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26]).reshape(-1, 1)
yi = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Crear el modelo de regresión lineal
regr = LinearRegression()

# Entrenar el modelo con los datos proporcionados
regr.fit(xi, yi)

# Predecir valores con el modelo entrenado
yi_pred = regr.predict(xi)

# Mostrar los coeficientes
pendiente = regr.coef_
intercept = regr.intercept_

# Calcular el error cuadrático medio y el coeficiente de determinación R^2
mse = mean_squared_error(yi, yi_pred)
r2 = r2_score(yi, yi_pred)

# Graficar los puntos y la recta de regresión
plt.scatter(xi, yi, color="black", label="Datos")
plt.plot(xi, yi_pred, color="blue", linewidth=3, label="Regresión")

# Etiquetas y leyenda
plt.xlabel("x_i")
plt.ylabel("y_i")
plt.title("Regresión lineal con datos de la tabla")
plt.legend()

# Mostrar la gráfica
plt.show()

print("Pendiente (slope):", pendiente)
print("Intercepto (intercept):", intercept)
print("Error cuadrático medio (MSE):", mse)
print("R-squared: (r2)", r2)
```

```python
# Cálculo realizado manualmente
import numpy as np
import matplotlib.pyplot as plt

# Datos obtenidos de la tabla
xi = np.array([2, 6, 8, 8, 12, 16, 20, 20, 22, 26])
yi = np.array([58, 105, 88, 118, 117, 137, 157, 169, 149, 202])

# Calcular la media de X y y
x_mean = np.mean(xi)
y_mean = np.mean(yi)

# Calcular la pendiente según formula
numerador = np.sum((xi - x_mean) * (yi - y_mean))
denominador = np.sum((xi - x_mean) ** 2)
pendiente = numerador / denominador

# Calcular el intercepto (intercept)
intercept = y_mean - pendiente * x_mean

# Usar la pendiente y el intercepto para calcular la línea de regresión
y_manual_pred = pendiente * xi + intercept

# Graficar datos originales
plt.scatter(xi, yi, color="black", label="Datos")

# Graficar línea de regresión calculada manualmente
plt.plot(xi, y_manual_pred, color="red", linewidth=3, label="Regresión (manual)")

# Etiquetas y leyenda
plt.xlabel("xi")
plt.ylabel("yi")
plt.title("Regresión Lineal Calculada 'Manualmente'")
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()

# Imprimir coeficientes
print("Pendiente (slope):", pendiente)
print("Intercepto (intercept):", intercept)
```
